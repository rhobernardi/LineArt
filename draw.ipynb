### Criando funções para abstrair

Para calcularmos o deslocamento de uma agulha pra outra e contar os pixels, precisamos achar o deslocamento de um ponto ao outro:

$dx = (x - x')$\
$dy = (y - y')$

$p = (x+dx, y+dy)$

$(x,y)$

$(R,G,B)$

```python
def reset_image(img):
    WHITE = [255,255,255]
    for row in range(0,img.shape[0]):
        for col in range(0,img.shape[1]):
            img[row, col] = WHITE

def draw_nail(img, row, col, radius=2, color=(255,0,0), thickness=1):
    center_coordinates = (col, row)
    img = cv2.circle(img, center_coordinates, radius, color, thickness)

def draw_line(img, start_point, end_point, color=(0,0,0), thickness=1):
    img = cv2.line(img, start_point, end_point, color, thickness)

# a funçao nao presta!!!!!!!! olhar pixels em azul


#até aqui consegui somente a contagem de pixels pretos numa reta, dada a agulha de saida e o de chegada
#def pixel_analysis(img, nr_nails, start_point, end_point):
#    tuplas = []
#    for n in range(nr_nails)
#        tuplas.append(pixel_count(img, start_point, end_point))
#    return 
```

### Gerando a janela e carregando as imagens

```python
animation_speed = 1

root = tk.Tk()
root.withdraw()
user32 = ctypes.windll.user32
screensize = user32.GetSystemMetrics(0), user32.GetSystemMetrics(1)

file_path = "batman.png" # filedialog.askopenfilename()
image = cv2.imread(file_path)
image = cv2.resize(image, (600, 600))
blank_image = np.zeros((image.shape[0],image.shape[1],3), dtype=np.uint8)
reset_image(blank_image)

display = cv2.resize(image, ((image.shape[0] * 2), image.shape[1]))
display[0:display.shape[0], 0:int(display.shape[1]/2)] = image
display[0:display.shape[0], int(display.shape[1]/2):display.shape[1]] = blank_image

center = [int(image.shape[0]/2), int(image.shape[1]/2)]
```

### Colocando os $n$ alfinetes na imagem.
Aqui é calculada a posição dos alfinetes usando coordenadas polares:

$x = r \cos(\theta)$\
$y = r \sin(\theta)$

$\theta \in [0,2\pi]$

```python
nr_nails = 200 # int(input('nails number: '))
angle = 360/nr_nails
nail_positions = []
radius = min(center[0],center[1])-1

for i in range(nr_nails):
    col = radius + int(radius*cos(-radians(angle*i)))
    row = radius + int(radius*sin(-radians(angle*i)))
    draw_nail(image, row, col)
    print(f"({row},{col})")
    # draw_nail(blank_image, row, col)
    display[0:display.shape[0], 0:int(display.shape[1]/2)] = image
    # display[0:display.shape[0], int(display.shape[1]/2):display.shape[1]] = blank_image
    nail_positions.append([row,col])
    cv2.imshow('Drawing lines', display)
    cv2.waitKey(animation_speed)
```
